```python
alpha = 0.1
```

Null Hypothesis: The 5 versions of the button are equally likely to receive clicks, and the observed differences are due to chance.means d = 0.
Alternative Hypothesis: The observed differences are not due to chance: there is at least one version that got so many more/much less clicks than the others.means, d!= 0.

```python
element_clicks_count = [42,53,21,38,45]
total_visits =  [10241,2689,2726,3142,2019]
total = pd.DataFrame([element_clicks_count, total_visits],
                           columns = ["Interact", "Connect","Learn","Help","Services"],
                           index = ["Clicks", "Visits"])
total.info()
```

```python
total.head()
```

```python
Interact_rate = total.Interact['Clicks']/total.Interact['Visits']
Connect_rate = total.Connect['Clicks']/total.Connect['Visits']
Learn_rate = total.Learn['Clicks']/total.Learn['Visits']
Help_rate = total.Help['Clicks']/total.Help['Visits']
Services_rate = total.Services['Clicks']/total.Services['Visits']
```

```python
rate = [Interact_rate,Connect_rate,Learn_rate,Help_rate,Services_rate]
element_rates = pd.DataFrame([rate],
                           columns = ["Interact", "Connect","Learn","Help","Services"])
element_rates
```

```python
from scipy import stats
chisq, pvalue, df, expected = stats.chi2_contingency(total)
```

```python
if pvalue > alpha:
  print("The p-value is larger than alpha.")
else:
  print("The p-value is smaller than alpha.")
```

Since the p-value is less than alpha, we can reject the Null Hypothesis.
meaning, we are 90% confident that all 5 versions are not same. The observed differences are not due to chance: there is at least one version that got so many clicks than the others
