**â€¢	What is the distribution of product prices across different categories.**

```python
price_category = (
    final.groupby(["category"])["unit_price"]
    .mean(["unit_price"]).round(2)
    .nlargest(30))
price_category
```

**Top ordered products**

```python
product_order_category = (
    final.groupby(["category"])["order_id"]
    .count()
    .nlargest(30)
)
product_order_category
```

```python
from matplotlib import pyplot as plt
product_order_category.plot(kind = "barh");

plt.title("Orders count of all product category");
plt.xlabel("Order Count");
plt.ylabel("Product Category");
```

**Top Revenue products**

```python
product_revenue_category = (
    final.groupby(["category"])["total_paid"]
    .sum(["total_paid"]).round(2)
    .nlargest(30))
product_revenue_category
```

```python
product_revenue_category.plot(kind = "barh");

plt.title("Revenue of all product category");
plt.xlabel("Revenue");
plt.ylabel("Product Category");
```
