**â€¢	How many products are being discounted.**

```python
discount_category = (
    final.groupby(["category"])["discount"]
    .mean(["discount"])
    .nlargest(30).round(2))
discount_category
```

```python
discount_category.plot(kind = 'bar');
```

**â€¢	What is the distribution of product prices across different categories.**

```python
price_category = (
    final.groupby(["category"])["unit_price"]
    .sum(["unit_price"])
    .nlargest(30))
price_category
```

```python
price_category.plot(kind = "bar");
```

```python
sns.relplot(data=final, 
            y="category", 
            x="unit_price",
            height=8, 
            aspect=1.5
            ).set(title='Distribution of products against unit_price');
```

**Date wise Average discount distribution.**

```python
discount_daily = final.groupby(pd.Grouper(key='created_date',freq='D'))['discount'].mean()
discount_daily
```

```python
sns.lineplot(data = discount_daily);
```
