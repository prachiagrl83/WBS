## **1. Import Libraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.model_selection import GridSearchCV
```

## **2. Data Reading**

```python
# reading mushroom train data
url = "https://drive.google.com/file/d/1u5GNc48zNBOhpMP3A2CCy8vmYyhPG0aK/view?usp=share_link"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
train = pd.read_csv(path)


# reading mushroom test data
url = "https://drive.google.com/file/d/1H5cuNzecAF1aJY-2ECxOTHSYGT-ABLuD/view?usp=share_link"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
test = pd.read_csv(path)
```

```python
# X and y creation for train data
X = train.drop('Id',axis=1).copy()
Y = X.pop("poisonous")
```

```python
#test = test.drop('Id',axis=1).copy()
```

## **2. Data Manipulation**

```python
#To check categorical values
for i in X:
    print(i, " : ", X[i].unique())
```

```python
# initialize
my_onehot = OneHotEncoder()

# fit
my_onehot.fit(X)
my_onehot.fit(test) #for test data
# transform
X_onehot = my_onehot.transform(X)
test_onehot = my_onehot.transform(test)#for test data
X_onehot
```

```python
test_onehot
```

```python
X_onehot.toarray()
X_new = pd.DataFrame.sparse.from_spmatrix(X_onehot)
X_new.head()
```

```python
#for test data
test_onehot.toarray()
test_new = pd.DataFrame.sparse.from_spmatrix(test_onehot)
test_new.head()
```

```python
colnames = my_onehot.get_feature_names_out(X.columns)
X_new.columns = colnames
X_new.head()
```

```python
#for test data
colnames = my_onehot.get_feature_names_out(test.columns)
test_new.columns = colnames
test_new.head()
```

## **3. Classification Methods**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_new, Y, test_size = 0.25, random_state = 6)
```

```python
#Decision Tree
tree = DecisionTreeClassifier()
tree.fit(X_new, Y) #or try with X_train and Y_train
tree_pred = tree.predict(test_new)
acc_tree = round(accuracy_score(Y_test, tree.predict(test_new))*100,2)
acc_tree
```

```python
#AUC Score
tree_predict_prob = tree.predict_proba(test_new)
fpr, tpr, thresholds = roc_curve(Y_test, tree_predict_prob[:,1])
tree_auc = auc(fpr, tpr)
print(tree_auc)
```

```python
#Logistic Regression Classification
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_new, Y)
lr_pred = lr.predict(test_new)
acc_lr = round(accuracy_score(Y_test, lr.predict(test_new))*100,2)
acc_lr
```

```python
lr_predict_prob = lr.predict_proba(test_new)
fpr, tpr, thresholds = roc_curve(Y_test, lr_predict_prob[:,1])
lr_auc = auc(fpr, tpr)
print(lr_auc)
```

```python
rf = RandomForestClassifier(n_estimators=7,criterion='gini',max_depth=5)
rf.fit(X_new, Y)
rf_pred = rf.predict(test_new)
acc_rf = round(accuracy_score(Y_test, rf.predict(test_new))*100,2)
acc_rf
```

```python
rf_predict_prob = rf.predict_proba(test_new)
fpr, tpr, thresholds = roc_curve(Y_test, rf_predict_prob[:,1])
rf_auc = auc(fpr, tpr)
print(rf_auc)
```

```python
knn = KNeighborsClassifier(n_neighbors=5,weights='distance')
knn.fit(X_new,Y)
knn_pred = knn.predict(test_new)
acc_knn = round(accuracy_score(Y_test, knn.predict(test_new))*100,2)
acc_knn
```

```python
#Checking performance our model with classification report.
from sklearn.metrics import classification_report
print(classification_report(Y_test, knn_pred))
```

```python
#Checking performance our model with ROC Score.
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test, knn_pred)
```

```python
knn_predict_prob = knn.predict_proba(test_new)
fpr, tpr, thresholds = roc_curve(Y_test, knn_predict_prob[:,1])
knn_auc = auc(fpr, tpr)
print(knn_auc)
```

```python
#Which is the best Model ?
results = pd.DataFrame({
    'Model': ['Linear Regression','Random Forest','Decision Tree','KNN'],
    'Score': [acc_lr, acc_rf, acc_tree,acc_knn]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df.head()
```
